Water potability, also known as the suitability of water for drinking, is essential for public health and safety. In this project, we aim to predict the classification of a water source as potable or not based on a range of water quality measurements. The dataset that will be analyzed contains water quality measurements including pH, sulfate concentration, turbidity, and mineral content, among others. Our primary question is straightforward: Can we effectively classify a water source as suitable for human consumption based on these different measurements? 

### Preliminary Exploratory Data Analysis

The following code demonstrates that the Kaggle Water Potability Dataset can be read into R. Note that the dataset was downloaded from Kaggle using the url listed in the code. The dataset was then pushed to our group's remote repository, which was then cloned in jupyter.

```r
#Kaggle URL
url <- "https://www.kaggle.com/datasets/uom190346a/water-quality-and-potability"

#Loading Relevant Libraries
library(tidyverse)
library(tidymodels)
library(RColorBrewer)

#Reading in the Water Potability Dataset
water_pot_data <- read_csv("water_potability.csv")

#A Glimps of the Data
glimpse(water_pot_data)
```

Now we ensure that the data is in tidy format and clean. Luckily the dataset is already tidy, but we should convert our classifier variable (potability) into a factor. We should also factor recode it so that instead of 0 and 1, the values are either potable or nonpotable.

```r
#Converting Potability into a factor and factor recoding for clarity

water_pot_data <- water_pot_data |>
    mutate(Potability = as_factor(Potability)) |>
    mutate(Potability = fct_recode(Potability, "Potable" = "1", "Unpotable" = "0"))

#A look at the modified row
water_modified_glimpse <- water_pot_data |>
    select(Potability) |>
    slice(1:5)
water_modified_glimpse
```

For the purposes of developing a predictive model, the clean and tidy data is split into training and testing sets with the training data having a 75% proportion.

```r
water_split <- initial_split(water_pot_data, prop = 0.75, strata = Potability)
water_train <- training(water_split)
water_test <- testing(water_split)
```

The following tables summarize useful properties of the training data. 

The first table illustrates the number of observations that are potable and nonpotable, respectively.

```r
water_potcount <- water_train |>
    group_by(Potability) |>
    summarize(count = n())

water_potcount
```

The second table summarizes the mean values for each of the predictor variables we plan to use for our analysis.

```r
water_means <- water_train |>
    select(ph, Solids, Hardness, Turbidity) |>
    map_df(mean, na.rm = TRUE)

water_means
```

The third table counts the number of observations in the training data that have NA as a value.

```r
water_na <- water_train |>
    summarize(NA_Count = sum(is.na(water_train)))
water_na
```

The distribution of our predictor variables is of particular interest given that they will inform the accuracy of our classification model. The following visualization summarizes the distribution of our chosen predictor variables through a facet grid of histograms.

```r
options(repr.plot.width = 16, repr.plot.height = 5)

modified_water <- water_train |>
    select(ph, Solids, Hardness, Turbidity) |>
    pivot_longer(cols = ph:Turbidity,
                 names_to = "Characteristic",
                 values_to = "Concentration")

characteristic_labels <- c(ph = "pH", 
                    Hardness = "Hardness", 
                    Solids = "Solids",
                    Turbidity = "Turbidity")

water_vis <- ggplot(modified_water, aes(x = Concentration, fill = Characteristic)) + 
    geom_histogram() + 
    facet_grid(cols = vars(Characteristic), scales = "free", 
              labeller = labeller(Characteristic = characteristic_labels))+
    xlab("Amount")+
    ylab("Count")+ 
    ggtitle("Relative Distribution of Water Quality Measurements") +
    theme(axis.text.x = element_text(angle = 90, hjust = 1),
         strip.text.y.left = element_text(angle = 0),
         text = element_text(size = 15),
         legend.position = "none") +
    scale_fill_brewer(palette = "Paired")

water_vis
```
